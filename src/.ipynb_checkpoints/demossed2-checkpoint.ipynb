```python
import cvxpy as cp
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
def obstacle_circle(img_maze, obstacles, radius):
    for center in obstacles:
        cv2.circle(img_maze, center, radius, (0, 255, 0), 1)
    return img_maze
```

```python
def obstacle_fxy(x_star_dot, y_star_dot, obstacle_no):
    radius_square = radius**2
    x_val = (robot_initial_point[0] + x_star_dot.sum() * dt - obstacles[obstacle_no][0]) #Iska variable name change karna padega
    y_val = (robot_initial_point[1] + y_star_dot.sum() * dt - obstacles[obstacle_no][1]) #Iska variable name change karna padega
    return -(pow(x_val,2) + pow(y_val,2) - radius_square)

def obstacle_grad_fxy(x_star_dot, y_star_dot, x_dot, y_dot, obstacle_no):
    x_val = (robot_initial_point[0] + x_star_dot.sum() * dt - obstacles[obstacle_no][0]) #Iska variable name change karna padega
    y_val = (robot_initial_point[1] + y_star_dot.sum() * dt - obstacles[obstacle_no][1]) #Iska variable name change karna padega
    ret_val1 = np.array([-2*dt*x_val] * len(x_star_dot))
    ret_val2 = np.array([-2*dt*y_val] * len(y_star_dot))
    term2 = ret_val1.reshape(1, -1) @ (x_dot - x_star_dot.reshape(-1, 1))
    term3 = ret_val2.reshape(1, -1) @ (y_dot - y_star_dot.reshape(-1, 1))
    return term2, term3
```

```python
def optimize(x_star_dot, y_star_dot, prev_loss):
    while 1:
        x_dot = cp.Variable(shape = (n, 1))
        y_dot = cp.Variable(shape = (n, 1))
        constraints = []
        for i in range(n):
            constraints.append(-v_max <= x_dot[i, 0])
            constraints.append(x_dot[i, 0] <= v_max)
            constraints.append(-v_max <= y_dot[i, 0])
            constraints.append(y_dot[i, 0] <= v_max)
            for j in range(len(obstacles)):
#                 grad_fxy = obstacle_grad_fxy(x_star_dot[:i+1], y_star_dot[:i+1], x_dot[:i+1], y_dot[:i+1], j)
                term1 = obstacle_fxy(x_star_dot[:i+1], y_star_dot[:i+1], j)
                term2, term3 = obstacle_grad_fxy(x_star_dot[:i+1],y_star_dot[:i+1],x_dot[:i+1],y_dot[:i+1],j)
                constraints.append(term1 + term2 + term3 <= 0)

        x_val_1 = (robot_initial_point[0] + cp.sum(x_dot) * dt - robot_goal_point[0]) #Iska variable name change
        y_val_1 = (robot_initial_point[1] + cp.sum(y_dot) * dt - robot_goal_point[1]) #Iska variable name change
        objective = cp.Minimize(pow(x_val_1,2) + pow(y_val_1,2))
        prob = cp.Problem(objective, constraints)
        #loss = prob.solve() Removed this line
        if np.abs(prev_loss - prob.solve()) <= 10:
            return x_dot, y_dot
            break
        prev_loss = prob.solve()
        x_star_dot, y_star_dot = x_dot.value.flatten(), y_dot.value.flatten()
```

```python
def output (img_maze, robot_initial_point, robot_goal_point, x_dot, y_dot,delta_t):
    x_dot_vals = (x_dot.value).reshape((-1,))
    y_dot_vals = (y_dot.value).reshape((-1,))
    x = np.zeros((x_dot.shape[0]+1,))
    y = np.zeros((y_dot.shape[0]+1,))
    x[0] = robot_initial_point[0]
    y[0] = robot_initial_point[1]
    for i in range(len(x_dot_vals)):
        x[i+1] = x[i]+x_dot_vals[i]*delta_t
        y[i+1] = y[i]+y_dot_vals[i]*delta_t
    
    x = np.round(x)
    y = np.round(y)
    x = x.astype(np.int)
    y = y.astype(np.int)
    
    for i in range(len(x)-1):
        cv2.line(img_maze, (x[i], y[i]), (x[i+1], y[i+1]), (0, 0, 255), 2)
        plt.imshow(img_maze)
        plt.show()
```

# Results

```python
#Variable initialization for robot
robot_goal_point = (30, 40)
robot_initial_point = (350, 370)
v_max = 20
n = 23
dt = 1

#Variable initialization for obstacle
img_maze = 255*np.ones(shape=(400, 400, 3)).astype(np.uint8)
obstacles = [(200, 200)]
radius = 100
img_maze = obstacle_circle(img_maze, obstacles, radius)

#Optimisation
x_star_dot = np.zeros(shape = n)
y_star_dot = np.zeros(shape = n)
prev_loss = np.inf
x_dot, y_dot = optimize(x_star_dot, y_star_dot, prev_loss)

#Output
output(img_maze, robot_initial_point, robot_goal_point, x_dot, y_dot,dt)
```
