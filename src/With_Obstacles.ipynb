```python
import cvxpy as cp
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
def letThereBeWorld(world_size,robot_initial_point,robot_goal_point):
    world = np.full((world_size[0],world_size[1],3), 255)
    world = world.astype(np.uint8)
    cv2.circle(world,robot_initial_point,5,(255,0,0),5) # Mark start point on the image
    cv2.circle(world,robot_goal_point,5,(0,255,0),5) # Mark goal point on the image
    return world

def addCircularObstacles(world, circular_obstacles):
    for i in range(len(circular_obstacles)):
        center = (circular_obstacles[i,0], circular_obstacles[i,1])
        radius = circular_obstacles[i,2]
        cv2.circle(world, center, radius, (0, 0, 0), -1)
    return world

def func(r_i,x_star_dot, y_star_dot, circular_obstacles,obstacle_no,delta_t):
    radius = circular_obstacles[obstacle_no][2]
    x_val = (r_i[0] + x_star_dot.sum() * delta_t - circular_obstacles[obstacle_no][0])
    y_val = (r_i[1] + y_star_dot.sum() * delta_t - circular_obstacles[obstacle_no][1])
    return -(pow(x_val,2) + pow(y_val,2) - radius**2)

def funcGradient(r_i,x_star_dot, y_star_dot, x_dot, y_dot, circular_obstacles,obstacle_no,delta_t):    
    x_val = (r_i[0] + x_star_dot.sum() * delta_t - circular_obstacles[obstacle_no][0]) 
    y_val = (r_i[1] + y_star_dot.sum() * delta_t - circular_obstacles[obstacle_no][1]) 
    term1 = np.array([-2*delta_t*x_val] * len(x_star_dot))
    term2 = np.array([-2*delta_t*y_val] * len(y_star_dot))
    ret_val1 = term1.reshape(1, -1) @ (x_dot - x_star_dot.reshape(-1, 1))
    ret_val2 = term2.reshape(1, -1) @ (y_dot - y_star_dot.reshape(-1, 1))
    return ret_val1, ret_val2

def getVels(robot_initial_point,robot_goal_point,v_min,v_max,delta_t,n,circular_obstacles):
    x_star_dot = np.zeros(shape = n)
    y_star_dot = np.zeros(shape = n)
    prev_loss = np.inf    
    while 1:
        x_dot = cp.Variable(shape = (n, 1))
        y_dot = cp.Variable(shape = (n, 1))
        constraints = []
        for i in range(n):
            constraints.append(v_min <= x_dot[i, 0])
            constraints.append(x_dot[i, 0] <= v_max)
            constraints.append(v_min <= y_dot[i, 0])
            constraints.append(y_dot[i, 0] <= v_max)
            
            for j in range(len(circular_obstacles)):
                term1 = func(robot_initial_point,x_star_dot[:i+1], y_star_dot[:i+1],\
                                 circular_obstacles,j,delta_t)
                term2,term3 = funcGradient(robot_initial_point,x_star_dot[:i+1], y_star_dot[:i+1],\
                                     x_dot[:i+1], y_dot[:i+1], circular_obstacles,j,delta_t)
                constraints.append(term1 + term2 + term3 <= 0)

        x_n = (robot_initial_point[0] + cp.sum(x_dot) * delta_t - robot_goal_point[0])
        y_n = (robot_initial_point[1] + cp.sum(y_dot) * delta_t - robot_goal_point[1])
        objective = cp.Minimize(pow(x_n,2) + pow(y_n,2))
        prob = cp.Problem(objective, constraints)
        if np.abs(prev_loss - prob.solve()) <= 10:
            return x_dot, y_dot
            break
        prev_loss = prob.solve()
        x_star_dot, y_star_dot = x_dot.value.flatten(), y_dot.value.flatten()
    return x_dot,y_dot

def getXY(robot_initial_point,x_dot,y_dot,delta_t):    
    x_dot_vals = (x_dot.value).reshape((-1,))
    y_dot_vals = (y_dot.value).reshape((-1,))
    x = np.zeros((x_dot.shape[0]+1,))
    y = np.zeros((y_dot.shape[0]+1,))
    x[0] = robot_initial_point[0]
    y[0] = robot_initial_point[1]
    for i in range(len(x_dot_vals)):
        x[i+1] = x[i]+x_dot_vals[i]*delta_t
        y[i+1] = y[i]+y_dot_vals[i]*delta_t
    
    x = np.round(x)
    y = np.round(y)
    x = x.astype(np.int)
    y = y.astype(np.int)
    return x,y


def markPath(x,y,world):
#     Draw the path
    for i in range(len(x)-1):
        cv2.line(world, (x[i], y[i]), (x[i+1], y[i+1]), (0, 0, 255), 5) # Draw the path
#         cv2.imwrite("../results/q1/frame_"+str(i+1)+".jpg",world) # save the image
        plt.imshow(world)
        plt.title("path after " + str(i+1) + " steps")
        plt.show()
    return None
```

# Results

```python
# Initialize variables
robot_initial_point = (75, 35)
robot_goal_point = (300, 350)
v_max = 20
v_min = -10
delta_t = 1
n = 20
# Create World
world_size = (400,400)
world = letThereBeWorld(world_size,robot_initial_point,robot_goal_point)

# Add obstacles to the world
circular_obstacles = [[200, 200,50]]
circular_obstacles = np.array(circular_obstacles)
world = addCircularObstacles(world,circular_obstacles)

# Show initial world
plt.imshow(world)
plt.title("Initial world")
plt.show()
# Solve the problem
x_dot, y_dot = getVels(robot_initial_point,robot_goal_point,v_min,v_max,delta_t,n,circular_obstacles)
x,y = getXY(robot_initial_point,x_dot,y_dot,delta_t)
markPath(x,y,world)
```
